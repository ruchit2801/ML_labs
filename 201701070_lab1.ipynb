# Question 3

```python
# Problem 3

# Import the dataset
dataset = pd.read_excel('/content/drive/My Drive/ML Lab Assignments/data for Problem1.xlsx')
y = dataset.y
x = dataset.x

# Preprocess dataset
y = y.to_numpy()
x = x.to_numpy()

#Visualising the dataset
plt.scatter(x, y, color='r')
plt.xlabel('x')
plt.ylabel('y')
plt.show()


# Calculate the loss function (Without normalising features)
L = (y[0] - (theta0 + theta1*x[0]))**2
for i in range(1,len(x)):
    L += (y[i] - (theta0 + theta1*x[i]))**2
L/=2*len(x)

# Plot loss function (without normalised loss function) 
ax = plt.axes(projection='3d')
ax.plot_surface(theta0, theta1, L)
ax.set_title('Surface plot (Without normalising features)')
ax.set_xlabel(r'$\theta_0$')
ax.set_ylabel(r'$\theta_1$')
ax.set_zlabel(r'$L(\Theta) = (y_i - (\theta_0 + \theta_1*x_i))$')
plt.show()


#Normalising features
x = (x - np.mean(x))/np.std(x)
y = (y - np.mean(y))/np.std(y)

# Calculate the loss function
L = (y[0] - (theta0 + theta1*x[0]))**2
for i in range(1,len(x)):
    L += (y[i] - (theta0 + theta1*x[i]))**2
L/=2*len(x)

# Plot loss function 
ax = plt.axes(projection='3d')
ax.plot_surface(theta0, theta1, L)
ax.set_title('Surface plot (After normalising features)')
ax.set_xlabel(r'$\theta_0$')
ax.set_ylabel(r'$\theta_1$')
ax.set_zlabel(r'$L(\Theta) = (y_i - (\theta_0 + \theta_1*x_i))$')
plt.show()
```

```python
# Obtaining theta that minimizes the loss function
theta0_min = theta[np.argmin(L, axis=1)[0]]
theta1_min = theta[np.argmin(L, axis=0)[0]]
print("theta0_min : {}".format(theta0_min))
print("theta1_min: {}".format(theta1_min))
print("L_min: {}".format(L[np.argmin(L, axis=0)[0]][np.argmin(L, axis=1)[0]]))
```
